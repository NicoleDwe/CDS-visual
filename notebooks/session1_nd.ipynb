## Image translation

```python
path_to_image = os.path.join("..", "data", "img", "trex.png")
image = cv2.imread(path_to_image)
jimshow(image, "Original")
```

__Shifting image down and right__

Translation matrix

shift 25 pixels on x-axis, 50 on y

```python
M = np.float64([[1, 0, 25], # x-axis
                [0, 1, 50]]) # y-axis
```

```python
width = image.shape[1]
height = image.shape[0]
```

```python
shifted = cv2.warpAffine(image, M, (width, height)) 
# using Affine trasnformation = preserves colinerity 
```

```python
jimshow(shifted, "Shifted down and right")
```

__Shifting image up and left__

```python
M = np.float64([[1, 0, -100], # x-axis
                [0, 1, -100]]) # y-axis
```

```python
shifted = cv2.warpAffine(image, M, (width, height)) 
# using Affine trasnformation = preserves colinerity
```

```python
jimshow(shifted, "Shifted up and left")
```

__What does this do?__

```python
M = np.float64([[1, 0, 0], # x-axis
                [0, 1, 100]]) # y-axis
```

```python
shifted = cv2.warpAffine(image, M, (width, height)) 
# using Affine trasnformation = preserves colinerity
```

```python
jimshow(shifted, "Shifted down")
```

__Writing a function for images__

```python
def translate(image, x, y): 
    # define translation matrix
    M = np.float64([[1, 0, x], # x-axis
                    [0, 1, y]]) # y-axis
    # perform translation on chosen image
    shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0])) 
    # return translated image
    
    return shifted
```

```python
shifted_image = translate(image, 0, 100)
```

```python
jimshow(shifted_image, "Shifted down")
```
